#Visualisasi Data Dengan Studi Kasus Yang Sudah Ditentukan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Players.csv')
```

**1. Visualisasi Pemain Berdasarkan Jenis Medali yang didapatkan**

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Gold'].value_counts().index, df['Gold'].value_counts().values, color='Orange')
plt.title('Jumlah Pemain Yang Mendapatkan Gold Medal')
plt.xlabel('Gold Medal')
plt.ylabel('Jumlah Pemain')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Silver'].value_counts().index, df['Silver'].value_counts().values, color='Blue')
plt.title('Jumlah Pemain Yang Mendapatkan Gold Silver')
plt.xlabel('Silver Medal')
plt.ylabel('Jumlah Pemain')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Bronze'].value_counts().index, df['Bronze'].value_counts().values, color='Brown')
plt.title('Jumlah Pemain Yang Mendapatkan Bronze')
plt.xlabel('Bronze Medal')
plt.ylabel('Jumlah Pemain')
plt.xticks(rotation=45)
plt.show()
```
